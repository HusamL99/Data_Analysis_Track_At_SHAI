# Basic Data Exploration


```python
import pandas as pd

df = pd.read_csv('/content/Salaries.csv')
df.head()
```

```python
# Number of rows and columns
num_rows, num_columns = df.shape
print(f'Number of rows: {num_rows}')
print(f'Number of columns: {num_columns}')
```

```python
# Data types of each column
data_types = df.dtypes
print('Data types of each column:')
print(data_types)
```

```python
# Check for missing values in each column
missing_values = df.isnull().sum()
print('Missing values in each column:')
print(missing_values)
```

# Descriptive Statistics

```python
mean_salary = df['TotalPayBenefits'].mean()
print(f'Mean Salary: {mean_salary:.2f}')
```

```python
median_salary = df['TotalPayBenefits'].median()
print(f'Median Salary: {median_salary:.2f}')
```

```python
mode_salary = df['TotalPayBenefits'].mode()
print(f'Mode Salary: {mode_salary.iloc[0]:.2f}')
```

```python
min_salary = df['TotalPayBenefits'].min()
max_salary = df['TotalPayBenefits'].max()

print(f'Minimum Salary: {min_salary:.2f}')
print(f'Maximum Salary: {max_salary:.2f}')
```

```python
salary_range = max_salary - min_salary
print(f'Salary Range: {salary_range:.2f}')
```

```python
std_dev_salary = df['TotalPayBenefits'].std()
print(f'Standard Deviation of Salary: {std_dev_salary:.2f}')

```

# Data Cleaning

```python
# Check for missing values in each column
missing_values = df.isnull().sum()
print('Missing values in each column:')
print(missing_values)
```

```python
df['BasePay'].fillna(df['BasePay'].median(), inplace=True)
print(df.head())

#Justification:
# 1. It is less sensitive to extreme values and provides a better measure of central tendency in the presence of outliers.
# 2. it is a reasonable choice when the missing values are randomly distributed and there is no clear pattern.
```

```python
df['OvertimePay'].fillna(df['OvertimePay'].median(), inplace=True)
df['OtherPay'].fillna(df['OtherPay'].median(), inplace=True)
```

```python

# Impute missing values in 'Benefits' with the median of non-missing values
benefits_median = df['Benefits'].median()
df['Benefits'].fillna(benefits_median, inplace=True)

# Print the first few rows of the DataFrame to verify the changes
print(df.head())
```

```python
df.drop(['Notes', 'Status'], axis=1, inplace=True)
# If a column contains mostly or entirely missing values, and the information is not critical for analysis, it's often better to drop the column rather than trying to impute or work with missing values.
```

```python
# Check for missing values in each column to double check

missing_values = df.isnull().sum()
print('Missing values in each column:')
print(missing_values)
```

# Basic Data Visualization


```python
import pandas as pd
df = globals()['df']
```

```python
import matplotlib.pyplot as plt

# Bar chart for the distribution of salaries


plt.figure(figsize=(10, 6))
plt.hist(df['TotalPay'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Salaries')
plt.xlabel('TotalPay')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Pie chart for the proportion of employees in different departments

department_counts = df['JobTitle'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(department_counts, labels=department_counts.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Proportion of Employees in Different Departments')
plt.show()
```

# Grouped Analysis

```python
# Group the data by the 'Department' column and calculate average salary for each department
department_avg_salary = df.groupby('JobTitle')['TotalPay'].mean()

# Display the summary statistics for each group
print(department_avg_salary)

# Bar chart for comparing average salaries across different departments
plt.figure(figsize=(12, 6))
department_avg_salary.sort_values().plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Average Salary by Department')
plt.xlabel('Department')
plt.ylabel('Average Salary')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# Simple Correlation Analysis

```python
# Identify any correlation between 'Salary' and another numerical column (e.g., 'YearsExperience')
correlation = df['TotalPay'].corr(df['Year'])

# Print the correlation coefficient
print(f'Correlation between Salary and YearsExperience: {correlation}')

# Scatter plot to visualize the relationship
plt.figure(figsize=(10, 6))
plt.scatter(df['Year'], df['TotalPay'], color='orange', alpha=0.5)
plt.title('Scatter Plot: Salary vs. YearsExperience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
